```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load_poverty_data():
    df_poverty = pd.read_csv("shared_data_read_only/Census-Data/ACSST5Y2022.S2201-Data.csv")
    
    df_poverty = df_poverty[df_poverty['NAME'].str.contains('Philadelphia')]
   
    df_poverty = df_poverty[df_poverty.index != 0]
    
    return df_poverty
```

```python
poverty_data = load_poverty_data()
```

```python
poverty_data = poverty_data.astype({"S2201_C03_021E": float, "S2201_C01_001E": float}, errors='ignore')
poverty_data = poverty_data[['GEO_ID', 'NAME', 'S2201_C03_021E', 'S2201_C01_001E']]
poverty_data['S2201_C01_001E'] = poverty_data['S2201_C01_001E'].replace(0, np.nan)
poverty_data['poverty Rate'] = np.where(poverty_data['S2201_C01_001E'] != 0, poverty_data['S2201_C03_021E'] / poverty_data['S2201_C01_001E'] * 100, np.nan)
poverty_data['Tract'] = poverty_data['GEO_ID'].str.split('US', expand=True)[1]
poverty_data.rename(columns={'Tract': 'Census Tract'}, inplace=True)

print("Poverty Data:")
poverty_data.head()
```

```python
zip_tract_table = pd.read_csv("shared_data_read_only/Geography-Mapping/ZIP_TRACT_122023.csv")
zip_tract_table['TRACT'] = zip_tract_table['TRACT'].astype(str)

tract_merged_df = pd.merge(poverty_data, zip_tract_table, left_on='Census Tract', right_on='TRACT', how='left')
tract_merged_df.drop('TRACT', axis=1, inplace=True)
tract_merged_df.rename(columns={'ZIP': 'Zip Code'}, inplace=True)

zip_pov_table = tract_merged_df[['Zip Code', 'NAME', 'S2201_C03_021E', 'S2201_C01_001E', 'poverty Rate']]

print("\nPoverty Data by ZIP Code:")
zip_pov_table.head()
```

```python
new_zip_pov_table_df = zip_pov_table.groupby('Zip Code').agg({'S2201_C03_021E': 'sum', 'S2201_C01_001E': 'sum', 'poverty Rate': 'sum'})
new_zip_pov_table_df['Weighted Avg Poverty Rate'] = new_zip_pov_table_df['S2201_C03_021E'] / new_zip_pov_table_df['S2201_C01_001E']
new_zip_pov_table_df.reset_index(inplace=True)

print("\nNew Poverty Data by ZIP Code:")
p25 = new_zip_pov_table_df['Weighted Avg Poverty Rate'].quantile(0.25)
p50 = new_zip_pov_table_df['Weighted Avg Poverty Rate'].quantile(0.50)
p75 = new_zip_pov_table_df['Weighted Avg Poverty Rate'].quantile(0.75)

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(new_zip_pov_table_df['Zip Code'], new_zip_pov_table_df['Weighted Avg Poverty Rate'], color='skyblue', label='Poverty Rate')
ax.axhline(y=p25, color='green', linestyle='--', label='25th percentile')
ax.axhline(y=p50, color='blue', linestyle='--', label='50th percentile (Median)')
ax.axhline(y=p75, color='orange', linestyle='--', label='75th percentile')
plt.title('Poverty Rate and Percentiles by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Poverty Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()

new_zip_pov_table_df.head()
```

```python
df_grad = pd.read_csv("shared_data_read_only/Education-Data/SDP_Graduation_Rates_School_S_2023-02-15.csv")
df_zip = pd.read_csv("shared_data_read_only/Education-Data/2023-2024 Master School List (20231003).csv")
```

```python
df_grad.replace('s', pd.NA, inplace=True)
df_grad['denom'] = pd.to_numeric(df_grad['denom'], errors='coerce')
df_grad['num'] = pd.to_numeric(df_grad['num'], errors='coerce')
percentiles = df_grad['num'].quantile([0.25, 0.5, 0.75])
df_grad['num_percentile'] = pd.cut(df_grad['num'], bins=[float('-inf'), *percentiles, float('inf')], labels=['<25th', '25th-50th', '50th-75th', '>75th'])
```

```python
merged_df = pd.merge(df_grad, df_zip[['ULCS Code', 'Zip Code']], left_on='schoolid_ulcs', right_on='ULCS Code', how='left')
merged_df.drop('ULCS Code', axis=1, inplace=True)
filtered_merged_df = merged_df[merged_df['subgroup'] == 'Economically Disadvantaged']
filtered_merged_df = filtered_merged_df[filtered_merged_df['rate_type'] == '4-Year Graduation Rate']
filtered_merged_df['year'] = filtered_merged_df['cohort'].str[:4].astype(int)
latest_years_df = filtered_merged_df.loc[filtered_merged_df.groupby('schoolid_ulcs')['year'].idxmax()]
latest_years_df.drop('year', axis=1, inplace=True)
```

```python
latest_years_df['num'] = pd.to_numeric(latest_years_df['num'], errors='coerce')
latest_years_df['denom'] = pd.to_numeric(latest_years_df['denom'], errors='coerce')
latest_years_df = latest_years_df.dropna(subset=['num', 'denom'])
zip_weight_df = latest_years_df.groupby('Zip Code').agg({'num': 'sum', 'denom': 'sum'})
zip_weight_df['Weighted Grad Rate'] = zip_weight_df['num'] / zip_weight_df['denom']
zip_weight_df.reset_index(inplace=True)
zip_weight_df = zip_weight_df[['Zip Code', 'Weighted Grad Rate']]
```

```python
plt.figure(figsize=(15, 5))
zip_weight_df.plot(x="Zip Code", y=["Weighted Grad Rate"], kind="bar", figsize=(15, 5))
plt.xlabel('Zip Code')
plt.ylabel('Weighted Grad Rate')
plt.title('Weighted Graduation Rate by Zip Code')
plt.show()
plt.savefig('weighted.png')
```

```python
merged_df = pd.merge(new_zip_pov_table_df, zip_weight_df, on='Zip Code', how='inner')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Weighted Avg Poverty Rate'], merged_df['Weighted Grad Rate'], color='skyblue')
coefficients = np.polyfit(merged_df['Weighted Avg Poverty Rate'], merged_df['Weighted Grad Rate'], 1)
poly_function = np.poly1d(coefficients)
plt.plot(merged_df['Weighted Avg Poverty Rate'], poly_function(merged_df['Weighted Avg Poverty Rate']), color='red')
plt.title('Graduation Rate vs. Poverty Rate')
plt.xlabel('Weighted Avg Poverty Rate')
plt.ylabel('Weighted Grad Rate')
plt.grid(True)
plt.legend(['Line of Best Fit', 'Data Points'])
plt.show()
plt.savefig('gradvpoverty.png')
correlation_coefficient = merged_df['Weighted Avg Poverty Rate'].corr(merged_df['Weighted Grad Rate'])
print("Correlation Coefficient:", correlation_coefficient)
```

```python
merged_df['Poverty Rate Z-Score'] = (merged_df['Weighted Avg Poverty Rate'] - merged_df['Weighted Avg Poverty Rate'].mean()) / merged_df['Weighted Avg Poverty Rate'].std()
merged_df['Grad Rate Z-Score'] = (merged_df['Weighted Grad Rate'] - merged_df['Weighted Grad Rate'].mean()) / merged_df['Weighted Grad Rate'].std()
merged_df['Combined Score'] = (merged_df['Poverty Rate Z-Score'] + merged_df['Grad Rate Z-Score']) / 2
most_help_needed = merged_df.sort_values(by='Combined Score', ascending=True)
print("Most Help Needed:")
print(most_help_needed[['Zip Code', 'Combined Score']].head())
```
